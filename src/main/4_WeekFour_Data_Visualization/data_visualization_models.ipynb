1. Line Chart
- Demonstration: The line chart below represents a sorted sequence of points connected by straight line segments. It's useful for showing trends over time or ordered categories.
- Explanation: Line charts are excellent for tracking changes over intervals, particularly when the changes are relatively small or when you have multiple data sets to compare.

```python
x = np.sort(np.random.rand(100))
y = 2 * x + 1  # Example function for demonstration

plt.figure(figsize=(8, 4))
plt.plot(x, y, label='Line Chart')
plt.title('Line Chart Demonstration')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```

2. Scatter Plot
- Demonstration: The scatter plot displays individual data points. Here, each dot represents a data point from the dataset.
- Explanation: Scatter plots are ideal for showing the relationship between two variables. They can reveal distribution trends and outliers effectively.

```python
x = np.random.rand(100)
y = 2 * x + 1 + np.random.normal(0, 0.1, 100)  # Adding some noise

plt.figure(figsize=(8, 4))
plt.scatter(x, y, label='Scatter Plot')
plt.title('Scatter Plot Demonstration')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```

3. Histogram
- Demonstration: The histogram shows the distribution of a numerical variable, dividing data into bins of equal width.
- Explanation: Histograms are used to summarize data that is measured on an interval scale (e.g., height, weight, age). They are great for understanding the distribution and central tendency of the data.

```python
data = np.random.normal(0, 1, 1000)  # Generating random data

plt.figure(figsize=(8, 4))
plt.hist(data, bins=20, label='Histogram')
plt.title('Histogram Demonstration')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

4. Box Plot
- Demonstration: The box plot displays the distribution of data based on a five-number summary: minimum, first quartile (Q1), median, third quartile (Q3), and maximum.
- Explanation: Box plots are useful for identifying outliers and understanding the spread and skewness of the data.

```python
df = pd.DataFrame({
    'Category': np.random.choice(['A', 'B', 'C'], 100),
    'Value': np.random.normal(0, 1, 100)
})

plt.figure(figsize=(8, 4))
sns.boxplot(data=df, x='Category', y='Value')
plt.title('Box Plot Demonstration')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()
```

5. Heatmap
- Demonstration: The heatmap visualizes data in a matrix format, where the colors represent data values.
- Explanation: Heatmaps are effective for displaying patterns in large datasets, correlations, or how two variables interact.

```python
# Creating a sample DataFrame with numeric values
np.random.seed(0)  # For reproducible output
data = np.random.rand(10, 10)  # Generate a 10x10 matrix of random numbers
df = pd.DataFrame(data, columns=[f'Feature{i}' for i in range(1, 11)])

# Compute the correlation matrix
corr = df.corr()

# Plotting the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Heatmap of Feature Correlations')
plt.show()
```

6. Linear Regression
- Demonstration: The linear regression model is represented by the red line, showing the relationship between the feature and target.
- Explanation: Linear regression is used to predict the value of a variable based on the value of another variable. The line of best fit represents the linear relationship between the two.

```python

x = np.random.rand(100, 1)
y = 2 * x + 1 + np.random.normal(0, 0.1, (100, 1))

model = LinearRegression()
model.fit(x, y)
y_pred = model.predict(x)

plt.figure(figsize=(8, 4))
plt.scatter(x, y, label='Data')
plt.plot(x, y_pred, color='red', label='Linear Regression')
plt.title('Linear Regression Demonstration')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```
